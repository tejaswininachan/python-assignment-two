```python
# Print numbers from 1 to 50 using for loop

for i in range(1, 51):
    print(i)
```

```python
# Print all even numbers between 1 and 100

for i in range(1, 101):
    if i % 2 == 0:
        print(i)
```

```python
# Find the sum of first N natural numbers

n = int(input("Enter N: "))

total = 0

for i in range(1, n + 1):
    total = total + i

print("Sum =", total)
```

```python
# Display multiplication table using for loop

num = int(input("Enter number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

```python
# Count how many numbers are divisible by 5

numbers = [10, 12, 15, 18, 20, 25, 33, 40]

count = 0

for i in numbers:
    if i % 5 == 0:
        count = count + 1

print("Numbers divisible by 5 =", count)
```
