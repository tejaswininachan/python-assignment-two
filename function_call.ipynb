```python
# Function to check prime number

def check_prime(num):

    if num <= 1:
        return "Not Prime"

    for i in range(2, num):

        if num % i == 0:
            return "Not Prime"

    return "Prime"

result = check_prime(7)

print(result)
```

```python
# Function to find largest number from list

def largest_number(numbers):

    return max(numbers)

nums = [10, 45, 67, 23, 89]

result = largest_number(nums)

print(result)
```

```python
# Function to count vowels in string

def count_vowels(text):

    vowels = "aeiouAEIOU"

    count = 0

    for char in text:

        if char in vowels:
            count = count + 1

    return count

result = count_vowels("Hello Python")

print(result)
```
